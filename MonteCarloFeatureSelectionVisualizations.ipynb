```python
ValueCounts = MonteCarlo_1['Feature Indices String'].value_counts()
ValueCounts
```

```python
FilteredValueCounts = ValueCounts[ValueCounts >= 25]
FilteredValueCounts
```

```python
print(list(FilteredValueCounts.index))
print(FilteredValueCounts.sum())
```

```python
Filtered_MonteCarlo = MonteCarlo_1[MonteCarlo_1["Feature Indices String"].isin(list(FilteredValueCounts.index))]
Filtered_MonteCarlo
```

```python
my_order = Filtered_MonteCarlo.groupby(by=["Feature Indices String"])["F1"].mean().sort_values().index
my_order
```

```python
fig, axs = plt.subplots(figsize=(20, 5))
# Box Plot Of MonteCarlo Simulations
sns.violinplot(
    Filtered_MonteCarlo, x="Feature Indices String", y="F1", hue="Feature Indices String",
    # whis=[0, 100], 
    width=.6, palette=sns.color_palette("husl", 20), inner="quart", density_norm="count", order=my_order
)

# Add in points to show each observation
sns.stripplot(Filtered_MonteCarlo, x="Feature Indices String", y="F1", size=5, color=".3", order=my_order)

# Tweak the visual presentation
# ax.xaxis.grid(True)

# plt.title("lalala")
axs.set_title("F1-Measure of 5000 Feature Selection Simulations, With Features Sets Selected At Least 25 Times Or More")
axs.set_xlabel("Selected Features")
axs.set_ylabel("F1 Measure")
# plt.title = "Accuracy Score of 5000 Feature Selection Simulations, With Features Selected At Least 50 Times Or More"
# plt.xlabel = "Selected Features"
# plt.ylabel = "Accuracy"
plt.xticks(rotation=90)
# plt.set(ylabel="Accuracy", xlabel="Selected Features")
# sns.despine(trim=True, left=True)
plt.savefig('/Users/Joseph/Desktop/ASD_MonteCarloFeatureSelections_5000_F1.png', bbox_inches='tight')
plt.show()
# plt.savefig('/Users/Joseph/Desktop/PD_MonteCarloFeatureSelections.png', bbox_inches='tight')
```
